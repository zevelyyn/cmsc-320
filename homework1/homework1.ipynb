```python
# import BeautifulSoup to scrape & selenium for webdriver
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
# set webdriver to Firefox
driver = webdriver.Firefox() 
driver.get("https://pypl.github.io/PYPL.html")
```

```python
# create a Beautiful Soup object to extract HTML
bs = BeautifulSoup(driver.page_source)
```

```python
# find all tables in the source code
table = bs.findAll("table")
```

```python
# view all table contents
table
```

```python
# view the first table (table we need to scrape data off)
print(table[0].prettify())
```

```python
# begin to extract programming languages
tab1 = table[0].findAll({"td"})
```

```python
tab1
```

```python
# find only programming language names
programmingLanguages = table[0].findAll('td', class_='')
```

```python
programmingLanguages
```

```python
programmingLanguages = []

#  go through each language and remove HTML
for language in table[0].findAll('td', class_=''):
    programmingLanguages.append(language.get_text())
```

```python
# remove last element since it isnt a language
programmingLanguages.pop()
```

```python
programmingLanguages
```

```python
# find the top five most used languages
topFive = []
i = 0

# go from the top and get the first five
while len(topFive) < 5:
    rank = tab1[i].get_text()
    name = tab1[i + 2].get_text()
    percentage = tab1[i + 3].get_text()

    topFive.append({
        'Rank': rank,
        'Name': name,
        'Percentage': percentage,
    })

    # increment i by 5
    i += 5
```

```python
topFive
```

```python
# find the number of shares for each programming language
shares = table[0].findAll('td', class_='right')
```

```python
shares
```

```python
# the other output is not what we wanted, remove .optCol to get only share percentages
shares = table[0].select('td.right:not(.optCol)')
```

```python
shares
```

```python
shares = []

#  go through each language and remove HTML
for percentage in table[0].select('td.right:not(.optCol)'):
    shares.append(float(percentage.text.strip('%')))
```

```python
shares
```

```python
# get the length of the list to find the median
len(shares)
```

```python
# since there is an odd number in the list, get the middle value
medianShares = shares[len(shares) // 2]
```

```python
medianShares
```

```python
# create a dictionary to store programming language & trend
popular = {}
for i in range(0, len(tab1), 5):
    name = tab1[i + 2].get_text()
    trend = float(tab1[i + 4].get_text().strip('%'))

    popular[name] = trend
```

```python
popular
```

```python
# reverse sort the dictionary by value
sortedTrends = sorted(popular.items(), key=lambda item: item[1], reverse=True)
```

```python
sortedTrends
```

```python
# get the top three languages that were the most popular in the last year
topThree = sortedTrends[:3]
```
