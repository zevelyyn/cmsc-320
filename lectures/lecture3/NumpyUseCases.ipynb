### Signal processing

```python
# Generate a sine wave signal
time = np.linspace(0, 1, 1000)
signal = np.sin(2 * np.pi * 5 * time)

# Plot the signal and its FFT
plt.subplot(3, 1, 1)
plt.plot(time, signal)
plt.title('Original Signal')

plt.subplot(3, 1, 3)
plt.magnitude_spectrum(signal, Fs=1)
plt.title('Frequency Spectrum')

plt.show()
```

### Data transformation

```python
# Handling missing values
data = np.array([[1, 2, 3], [4, np.nan, 6], [7, 8, 9]])
cleaned_data = np.nan_to_num(data, nan=0)

# Standardize features
mean = np.mean(data, axis=0)
std_dev = np.std(data, axis=0)
standardized_data = (data - mean) / std_dev
```

### Hypothesis testing

```python
# Generate two random samples
sample1 = np.random.randn(100)
sample2 = np.random.randn(120)

# Perform independent two-sample t-test
t_stat, p_value = ttest_ind(sample1, sample2)
```
