```python
import requests
import csv
from io import StringIO
```

```python
# URL of a sample CSV file on GitHub
csv_url  = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

# Make a GET request to retrieve the CSV file
response = requests.get(csv_url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Decode the content of the response as a string
    csv_content = response.text

    # Use the csv module to read the CSV content
    csv_reader = csv.reader(StringIO(csv_content))

    # Iterate through the rows and print them
    for row in csv_reader:
        print(row)
else:
    # If the request was not successful, print an error message
    print("Error:", response.status_code)
    print("Response:", response.text)
```

```python
# Example URL of a public JSON API (JSONPlaceholder)
json_url = "https://jsonplaceholder.typicode.com/todos/1"

# Make a GET request to retrieve JSON data
response = requests.get(json_url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the JSON response
    json_data = response.json()

    # Print the JSON data
    print("Example - Retrieve JSON Data:")
    print(json_data)

    # Access specific fields in the JSON data
    print("\nAccessing Specific Fields:")
    print("User ID:", json_data['userId'])
    print("Title:", json_data['title'])
    print("Completed:", json_data['completed'])
else:
    # If the request was not successful, print an error message
    print("Error:", response.status_code)
    print("Response:", response.text)
```
