```python
country=[]
totalCases =[]
totalDeaths =[]

for row in tab1[9:232]: 
    try:
        col1= row.findAll("td")[1]      
        if col1 != None:
            country.append(col1)
#     print(col1.get_text())
        col2= row.findAll("td")[2]
        if col2 != None:
            totalCases.append(col2)
#     print(col2.get_text())
        col3= row.findAll("td")[4]
        if col3 != None:
            totalDeaths.append(col3)
#     print(col3.get_text())
   
    except:
        pass

   
    
    
```

```python
len(country)
```

```python
len(totalCases)
```

```python
len(totalDeaths)
```

```python
data=[]
for c, tc, td  in zip(country, totalCases, totalDeaths ):
    data.append([c.text, tc.text, td.text])
    print(c.text, tc.text, td.text)
```

```python
import csv
with open('cvCases.csv','w') as f:
    writer = csv.writer(f)
    writer.writerows(data)
    
    
```
